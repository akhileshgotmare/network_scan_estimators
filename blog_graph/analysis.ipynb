```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import random

path = 'blog_data/blogs_data_edge_list.txt'
edges_data = np.loadtxt(path, dtype = int)

edges_data[:,0] -= 1
edges_data[:,1] -= 1

N = np.max(edges_data)

G = nx.Graph()
G.add_nodes_from([n for n in range(N+1)])
G.add_edges_from(edges_data[edges_data[:,2] == 1] [:,0:2])
N = np.max(G.nodes()) + 1 

connected_comp_sub = nx.connected_component_subgraphs(G)
remove_list = list([])
for g in connected_comp_sub:
    if len(list(nx.nodes(g))) <= 2:
        remove_list = remove_list + (list(nx.nodes(g)))
for r in remove_list:
    G.remove_node(r)

```

```python
color1 = 'orange'
color2 = 'blue'
color3 = 'green'
color4 = 'red'
color5 = 'gray'

plt.rcParams["figure.figsize"] = [16,5]
from scipy.stats import norm
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

(mu,sigma) = norm.fit(a_comb_50)
n, bins, patches = plt.hist(a_comb_50, 60, facecolor=color1, alpha=0.5,label = 'k = 50',normed=1)
plt.axvline(2, color='k', linestyle='solid', linewidth=5,label = 'true value of $a$')
y = mlab.normpdf( bins, mu, sigma)
plt.xlim(1.4,2.4)
l = plt.plot(bins, y, 'orange',linestyle='dashed', linewidth=2)
plt.grid(True)
#plt.show()

(mu,sigma) = norm.fit(a_comb_100)
n, bins, patches = plt.hist(a_comb_100, 60,  facecolor=color2, alpha=0.5,label = 'k = 100',normed=1)#normed=1,
#plt.axvline(a_comb_100.mean(), color=color2, linestyle='solid', linewidth=5,label = 'mean of estimates $\hat{a}$ with k = 100')
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'b--', linewidth=2)
plt.grid(True)

(mu,sigma) = norm.fit(a_comb_150)
#plt.axvline(a_comb_150.mean(), color=color3, linestyle='solid', linewidth=5,label = 'mean of estimates $\hat{a}$ with k = 150')
n, bins, patches = plt.hist(a_comb_150, 60, facecolor=color3, alpha=0.5,label = 'k = 150',normed=1)
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, color = color3,linestyle = 'dashed', linewidth=2)
plt.grid(True)

(mu,sigma) = norm.fit(a_comb_200)
#plt.axvline(a_comb_200.mean(), color = color4, linestyle='solid', linewidth=5,label = 'mean of estimates $\hat{a}$ with k = 200')
n, bins, patches = plt.hist(a_comb_200, 60, facecolor=color4, alpha=0.5,label = 'k = 200',normed=1)
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y,color = color4 ,linestyle = 'dashed', linewidth=2)
plt.grid(True)

(mu,sigma) = norm.fit(a_comb_300)
#plt.axvline(a_comb_300.mean(), color=color5, linestyle='solid', linewidth=5,label = 'mean of estimates $\hat{a}$ with k = 300')
n, bins, patches = plt.hist(a_comb_300, 60, facecolor=color5, alpha=0.5,label = 'k = 300',normed=1)
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, color5,linestyle = 'dashed', linewidth=2)
plt.grid(True)
plt.legend(loc='upper left')

plt.annotate('$k = 50$', rotation=90, xy=(a_comb_50.mean(), 0), xytext=(a_comb_50.mean(), -1),
            arrowprops=dict(facecolor='orange', shrink=0.05),
            )

plt.annotate('$k = 100$', rotation=90, xy=(a_comb_100.mean(), 0), xytext=(a_comb_100.mean(), -1),
            arrowprops=dict(facecolor='blue', shrink=0.05),
            )


plt.annotate('$k = 150$', rotation=90, xy=(a_comb_150.mean(), 0), xytext=(a_comb_150.mean(), -1),
            arrowprops=dict(facecolor='green', shrink=0.05),
            )


plt.annotate('$k = 200$', rotation=90, xy=(a_comb_200.mean(), 0), xytext=(a_comb_200.mean(), -1),
            arrowprops=dict(facecolor='red', shrink=0.05),
            )
plt.annotate('${k = 300}$', rotation=90, xy=(a_comb_300.mean(), 0), xytext=(a_comb_300.mean(), -1),
            arrowprops=dict(facecolor='gray', shrink=0.05),
            )
plt.xlabel('estimate value')
plt.ylabel('Relative frequency')

#$\mu_{k = 300}$
plt.tight_layout()

plt.savefig('blog_hist_results.png',format = 'png', dpi=1000,  bbox_inches='tight')
plt.savefig('blog_hist_results.pdf', format='pdf', dpi=1000 , bbox_inches='tight') #, pdf, ps, eps )
plt.savefig('blog_hist_results.eps', format='eps', dpi=1000, bbox_inches='tight')
plt.show()

```
